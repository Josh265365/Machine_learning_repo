```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("winequality.csv").dropna()
df
```

```python
df['type'].value_counts() #prints the distribution of data across classes
```

```python
x = df.drop('type', axis=1)
y = df['type']
trainX, testX, trainY, testY = train_test_split(x, y, test_size = 0.2)
```

```python
#Scaling
sc=StandardScaler()
scaler = sc.fit(trainX)
trainX_scaled = scaler.transform(trainX)
testX_scaled = scaler.transform(testX)
```

```python
#Training MLP classiffier
mlp_clf = MLPClassifier(hidden_layer_sizes=(150,100,50),
max_iter = 300,activation = 'relu',
solver = 'adam')
mlp_clf.fit(trainX_scaled, trainY)
```

```python
y_pred = mlp_clf.predict(testX_scaled)
print('Accuracy: {:.2f}'.format(accuracy_score(testY, y_pred)))

from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(testY, y_pred)
print(classification_report(testY, y_pred))#F1 scores, precision recall
```

```python
#setting up grid
param_grid = {
'hidden_layer_sizes': [(50,40,50), (70,80,40), (80,50,30)],
'max_iter': [50, 20, 20],
'activation': ['tanh', 'relu'],
'solver': ['sgd', 'adam'],
'alpha': [0.0001, 0.05],
'learning_rate': ['constant','adaptive'],
}
```

```python
grid = GridSearchCV(mlp_clf, param_grid, n_jobs= -1, cv=5)
grid.fit(trainX_scaled, trainY)
print(grid.best_params_)
```
