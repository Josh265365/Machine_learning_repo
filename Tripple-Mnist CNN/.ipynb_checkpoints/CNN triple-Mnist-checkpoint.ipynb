```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
import cv2
import os
from sklearn.preprocessing import LabelEncoder

from pathlib import Path 
import random
from sklearn.tree import DecisionTreeClassifier
import tensorflow as tf
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, learning_curve
from tensorflow.keras import layers, models
from keras.layers import Dense, Input, Dropout, GlobalAveragePooling2D, Flatten, Conv2D, BatchNormalization, Activation, MaxPooling2D
```

TASK 1

```python
##LOAD IMAGES AND LABELS FUNCTIONS
##FOLLOW THE GIVEN PATH AND EXTRACT ALL "png" FILES
##READ ALL IMAGES AND APPEND THEM TO A LIST
###RETURN THE IMAGES AND LABELS IN A LIST
def load_images_and_labels(data_dir):
    images = []
    labels = []

    for root, _, files in os.walk(data_dir):
        for fname in files:
            if Path(fname).suffix == '.png':
                image_path = os.path.join(root, fname)
                image = cv2.imread(image_path)
                label = os.path.basename(root)
                images.append(image)
                labels.append(label)

    return images, labels
```

```python
#STORE DATA PATH IN VARIABLES
data_dir = 'triple_mnist/'
train_dir = os.path.join(data_dir, 'train')
val_dir = os.path.join(data_dir, 'val')
test_dir = os.path.join(data_dir, 'test')

#LOAD TRAINING IMAGES AND LABELS USING FUNCTION
train_images, train_labels = load_images_and_labels(train_dir)
num_train_images = len(train_images)
num_train_categories = len(set(train_labels))

print(f"Training: {num_train_images} images loaded with {num_train_categories} categories")

#LOAD VALIDATION IMAGES AND LABELS USING FUNCTION
val_images, val_labels = load_images_and_labels(val_dir)
num_val_images = len(val_images)
num_val_categories = len(set(val_labels))

print(f"Validation: {num_val_images} images loaded with {num_val_categories} categories")

#LOAD TEST IMAGES AND LABELS USING FUNCTION
test_images, test_labels = load_images_and_labels(test_dir)
num_test_images = len(test_images)
num_test_categories = len(set(test_labels))

print(f"Test: {num_test_images} images loaded with {num_test_categories} categories")
```

```python
# SELECT 5 RANDOM IMAGES FROM THE TRAINING DATASET
random_indices = random.sample(range(num_train_images), 6)


fig, axes = plt.subplots(2, 3, figsize=(10, 7))

# RESHAPE AXES TO FLAT 1D ARRAY
axes = axes.flatten()

# VISUALIZE RANDOM SAMPLES FROM THE TRAINING IMAGES
for i, index in enumerate(random_indices):
    image = train_images[index]
    label = train_labels[index]

    axes[i].imshow(image)
    axes[i].set_title(f"Label: {label}")
    axes[i].axis('off')

#REMOVE ANY EMPTY SLOTS
for j in range(len(random_indices), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

TASK 2

```python
# CONVERT THE LIST OF TRAINING AND TESTING IMAGES INTO NUMPY ARRAY
train_images_array = np.array(train_images)
train_labels_array = np.array(train_labels)

val_images_array = np.array(val_images)
val_labels_array= np.array(val_labels)

test_images_array = np.array(test_images)
test_labels_array= np.array(test_labels)

#CONVERTING THE RGB INPUT IMAGES TO GRAY AND NORMALIZING
train_images_gray = np.array([cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) / 255.0 for img in train_images_array])
test_images_gray = np.array([cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) / 255.0 for img in test_images_array])
val_images_gray = np.array([cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) /255.0 for img in val_images_array])

#FLATTENING THE GRAYSCALE IMAGES TO A SINGLE VECTOR (84X84 = 7056)
train_images_flat = train_images_gray.reshape(train_images_gray.shape[0], -1)
test_images_flat = test_images_gray.reshape(test_images_gray .shape[0], -1)
val_images_flat = val_images_gray.reshape(val_images_gray.shape[0], -1)

print('train images shape: ', train_images_flat .shape)
print('train labels shape:', train_labels_array.shape)

print('test images shape:', test_images_flat.shape)
print('test labels shape: ', test_labels_array.shape)

print('val images shape:', val_images_flat.shape)
print('val labels shape: ', val_labels_array.shape)
```

```python
#LABEL ENCODE THE TRAIN LABELS (DECISION TREE LABELS)
label_encoder = LabelEncoder()

train_labels_encoded = label_encoder.fit_transform(train_labels_array)
val_labels_encoded = label_encoder.fit_transform(val_labels_array)
test_labels_encoded = label_encoder.fit_transform(test_labels_array)
```

```python
#CONVERT EACH CHARACTER TO AN INT
#ONE HOT ENCODING LABELS 
###RETURN A 2D ARRAY OF ENCODED LABELS
def process_labels(labels):
    labels_split = [[int(digit) for digit in label] for label in labels]
    labels_split = np.array(labels_split)
    labels_onehot = [to_categorical(label, num_classes=10) for label in labels_split]
    return np.array(labels_onehot).reshape(len(labels_split), -1)

train_labels_enc = process_labels(train_labels_array)
test_labels_enc = process_labels(test_labels_array)
val_labels_enc = process_labels(val_labels_array)
```

```python
train_labels_enc.shape
```

```python
#TRAINING A DECISION TREE 

dt_clf = DecisionTreeClassifier(random_state=42)

# PARAMETER GRID FOR PARAMETER TUNNING
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 4, 5],
    'min_samples_split':[2,5,10],
}

# GRID SEARCH WITH K FOLD 2
grid_search = GridSearchCV(dt_clf, param_grid, cv=2, scoring='accuracy', n_jobs=2)
grid_search.fit(train_images_flat, train_labels_array)

# GETTING THE BEST PARAMETERS
best_dt_clf = grid_search.best_estimator_

print("Best Hyperparameters:", grid_search.best_params_)
```

```python
# Evaluate on the validation set
y_pred_val = best_dt_clf.predict(val_images_flat)
accuracy_val = accuracy_score(val_labels_array, y_pred_val)
f1_val = f1_score(val_labels_array, y_pred_val, average='weighted')
conf_matrix_val = confusion_matrix(val_labels_array, y_pred_val)

print("Validation Set Metrics:")
print("Accuracy:", accuracy_val)
print("F1 Score:", f1_val)
print("Confusion Matrix:\n", conf_matrix_val)
```

```python
# Evaluate on the test set
y_pred_test = best_dt_clf.predict(test_images_flat)
accuracy_test = accuracy_score(test_labels_array, y_pred_test)
f1_test = f1_score(test_labels_array, y_pred_test, average='weighted')
conf_matrix_test = confusion_matrix(test_labels_array, y_pred_test)

print("\nTest Set Metrics:")
print("Accuracy:", accuracy_test)
print("F1 Score:", f1_test)
print("Confusion Matrix:\n", conf_matrix_test)
```

```python
#SPLIT THE INPUT IMAGES INTO 3 PARTS
##CONVERT THEM TO GRAY SCALE AND NORMALIZE
#RETURN THE IMAGE WITH 3 PARTS
num_parts = 3

def split_image(image):
     height, width, channels = image.shape
     part_width = width // num_parts
     image_parts = [cv2.cvtColor(image[:, i * part_width: (i + 1) * part_width, :], cv2.COLOR_BGR2GRAY) for i in range(num_parts)]
     #image_parts = [image[:, i * part_width: (i + 1) * part_width, :] for i in range(num_parts)]
     
    
     return image_parts


# APPLY split_image FUNCTION TO EACH IMAGE 
train_images_parts = np.array([split_image(image) for image in train_images_array])
test_images_parts = np.array([split_image(image) for image in test_images_array])
val_images_parts = np.array([split_image(image) for image in val_images_array])
```

```python
#SPLIT THE LABEL IMAGES INTO 3 PARTS
#CONVERT EACH CHARACTER TO AN INT
###RETURN SPLIT LABELS
def split_label(label):
    
    label_parts = [int(digit) for digit in label]
    return label_parts

# Apply the split_label function to each label in the dataset
train_labels_parts = np.array([split_label(label) for label in train_labels_array])
test_labels_parts = np.array([split_label(label) for label in test_labels_array])
val_labels_parts = np.array([split_label(label) for label in val_labels_array])
```

```python
test_images_parts.shape
```

```python
#NORMALIZING TRAINING IMAGES FOR CNN
train_images_parts = train_images_parts.astype('float32')/255.0
test_images_parts = test_images_parts.astype('float32')/255.0
val_images_parts = val_images_parts.astype('float32')/255.0
```

```python
# Define height and part_width
height = train_images_parts.shape[2]
part_width = train_images_parts.shape[3]
#channels = train_images_parts.shape[4]

# RESHAPING INPUT IMAGES FOR THE CNN
train_images_parts_cnn = train_images_parts.reshape(train_images_parts.shape[0] * num_parts, height, part_width, 1)
test_images_parts_cnn = test_images_parts.reshape(test_images_parts.shape[0] * num_parts, height, part_width,1)
val_images_parts_cnn = val_images_parts.reshape(val_images_parts.shape[0] * num_parts, height, part_width, 1)
```

```python
def replicate_labels(labels, num_parts=3):
    replicated_labels = []
    for label in labels:
        for digit in label:
            digit_label = to_categorical(int(digit), num_classes=10)
            
            replicated_labels.append(digit_label)
    return np.array(replicated_labels)

# Apply the replicate_labels function to each label in the dataset
train_labels_replicated = replicate_labels(train_labels_parts)
test_labels_replicated = replicate_labels(test_labels_parts)
val_labels_replicated = replicate_labels(val_labels_parts)
```

```python
train_labels_replicated.shape
```

```python
from tensorflow.keras import models, layers
model = models.Sequential()

#model.add(layers.Conv2D(256, (3, 3), activation='relu', input_shape=(height, part_width, channels)))
model.add(layers.Conv2D(512, (3, 3), activation='relu', input_shape=(height, part_width , 1)))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.25))

model.add(layers.Conv2D(256, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.25))

model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.25))


model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.5))

model.add(layers.Dense(10, activation='softmax')) 

model.summary()
```

```python

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
#model.fit(train_images_parts_cnn, train_labels_enc, epochs=10, validation_data=(val_images_parts_cnn, val_labels_enc))

history = model.fit(train_images_parts_cnn, train_labels_replicated.astype('float32'), epochs=10, validation_data=(val_images_parts_cnn, val_labels_replicated.astype('float32')))

#model.fit(train_images_parts_cnn, train_labels_enc, epochs=10, validation_data=(val_images_parts_cnn, val_labels_enc))
```

```python
from tensorflow.keras import models, layers
model = models.Sequential()

model.add(layers.Conv2D(512, (3, 3), activation='relu', input_shape=(height, part_width , 1)))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.25))

model.add(layers.Conv2D(256, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.25))

model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.25))



model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.5))



head_1 = layers.Dense(10, activation='softmax', name='digit1')(model.output)
head_2 = layers.Dense(10, activation='softmax', name='digit2')(model.output)
head_3 = layers.Dense(10, activation='softmax', name='digit3')(model.output)

model = models.Model(inputs=model.input, outputs=[head_1, head_2, head_3])



model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(train_images_parts_cnn, train_labels_replicated.astype('float32'), epochs=5,batch_size=128, validation_data=(val_images_parts_cnn, val_labels_replicated.astype('float32')))
```
